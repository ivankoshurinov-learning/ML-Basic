## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
number = input("Введите пятизначное число: ")

if number.isdigit() and len(number) == 5:
    first_dig = number[0]
    last_dig = number[4]
    middle_digs = number[1:4][::-1]
    result = first_dig + middle_digs + last_dig
    print("Результат:", result)

else:
    print("Ошибка, вы ввели не пятизначное число")
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
days = int(input("Введите, сколько дней осталось до отпуска?: "))

weekends = 0

for day in range(1, days + 1):
    day_of_week = day % 7
    if day_of_week == 0:
        day_of_week = 7

    if day_of_week == 6 or day_of_week == 7:
        weekends += 1

print(f"До отпуска осталось {weekends} выходных дней")
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
choco_length = int(input("Длина шоколадки (в кусках): "))
choco_width = int(input("Ширина шоколадки (в кусках): "))
choco_desired = int(input("Размер куска, который хотите отломить: "))

can_break = False

if choco_desired % choco_length == 0:
    rows = choco_desired // choco_length
    if rows <= choco_width:
        can_break = True

elif choco_desired % choco_width == 0:
    columns = choco_desired // choco_width
    if columns <= choco_length:
        can_break = True

print(f"{choco_length}, {choco_width}, {choco_desired} -> {can_break}")
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
number_arab = int(input("Введите целое положительное число (1–3999): "))

if number_arab <= 0 or number_arab > 3999:
    print("Ошибка! Для конвертации в римское число должно быть от 1 до 3999")

else:
    number_roman = ""
    values = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
    symbols = ["M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]

    for i in range(len(values)):
        while number_arab >= values[i]:
            number_roman += symbols[i]
            number_arab -= values[i]

    print(number_roman)
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
user_input = input("Введите число для проверки: ")

if user_input.startswith("-"):
    num = user_input[1:]
else:
    num = user_input

dots_count = num.count(".")

if dots_count > 1:
    print(False)
else:
    temp = num.replace(".", "")

    if temp.isdigit() and len(temp) > 0:
        print(True)
    else:
        print(False)
```
