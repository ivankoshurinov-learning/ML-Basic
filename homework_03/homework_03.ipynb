## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
# в аргументе force_format указать:
# 'snake'  - это вернуть snake_case
# 'pascal' - это вернуть PascalCase

def convert_case(text: str, force_format: str | None = None) -> str:
    if "_" in text:
        current_format = "snake"
    else:
        current_format = "pascal"

    if force_format == "pascal":
        words = text.split("_")
        return "".join(word.capitalize() for word in words)

    if force_format == "snake":
        result = ""
        for letter in text:
            if letter.isupper():
                if result != "":
                    result += "_"
                result += letter.lower()
            else:
                result += letter
        return result

    if current_format == "snake":
        words = text.split("_")
        return "".join(word.capitalize() for word in words)

    else:
        result = ""
        for letter in text:
            if letter.isupper():
                if result != "":
                    result += "_"
                result += letter.lower()
            else:
                result += letter

        return result

print(f"otus_course -> {convert_case('otus_course')}")
print(f"PythonIsTheBest -> {convert_case('PythonIsTheBest')}")
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
def is_leap_year(year: int) -> bool:
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

def is_valid_date(date_str: str) -> bool:
# Подразумеваемый формат даты DD.MM.YYYY
# Проверяем формат и валидность цифровых значений

    parts = date_str.split(".")

    if len(parts) != 3:
        return False

    day, month, year = parts

    if not (day.isdigit() and month.isdigit() and year.isdigit()):
        return False

    day = int(day)
    month = int(month)
    year = int(year)

    if month < 1 or month > 12:
        return False

    days_in_month = {
        1: 31, 2: 28, 3: 31, 4: 30,
        5: 31, 6: 30, 7: 31, 8: 31,
        9: 30, 10: 31, 11: 30, 12: 31
    }

    max_days = 29 if month == 2 and is_leap_year(year) else days_in_month[month]

    return 1 <= day <= max_days

examples = [
    "29.02.2000",
    "29.02.2001",
    "31.04.1962"
]

for date in examples:
    print(f"{date} -> {is_valid_date(date)}")
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
def is_prime(number: int) -> bool:

    if number < 2:
        return False

    for i in range(2, number):
        if number % i == 0:
            return False

    return True

examples = [17, 20, 23]

for num in examples:
    print(f"{num} -> {is_prime(num)}")
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
def norm_name(value: str) -> str | None:
    if not value.isalpha():
        return None
    return value.capitalize()

def validate_age(value: str) -> int | None:
    if not value.isdigit():
        return None
    age = int(value)
    return age if 18 <= age <= 60 else None

def validate_id(value: str, users: dict) -> str | None:
    if not value.isdigit():
        return None
    user_id = value.zfill(8)
    if user_id in users:
        return None
    return user_id

def print_users(users: dict) -> None:

    print("\nСписок пользователей")
    print("-" * 40)
    print(f"{'ID':8} | {'Имя':10} | {'Фамилия':10} | {'Возраст'}")
    print("-" * 40)
    for user_id, (name, surname, age) in users.items():
        print(f"{user_id:8} | {name:10} | {surname:10} | {age}")
    print("-" * 40)

def main() -> None:
    users: dict[str, tuple[str, str, int]] = {}
    while True:
        name_input = input("Введите имя: ")
        if name_input == "":
            break

        surname_input = input("Введите фамилию: ")
        age_input = input("Введите возраст: ")
        id_input = input("Введите ID: ")

        name = norm_name(name_input)
        surname = norm_name(surname_input)
        age = validate_age(age_input)
        user_id = validate_id(id_input, users)

        if name is None or surname is None:
            print("Ошибка: имя и фамилия должны содержать только буквы\n")
            continue

        if age is None:
            print("Ошибка: возраст должен быть от 18 до 60\n")
            continue

        if user_id is None:
            print("Ошибка: ID должен быть уникальным числом\n")
            continue

        users[user_id] = (name, surname, age)

    print_users(users)

main()
```
